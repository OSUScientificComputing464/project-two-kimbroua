### c) Split data into training and test set.  Evaluate Linear Regression (OLS), Ridge, Lasso and ElasticNet using cross-validation with the default parameters.  Does scaling the data with StandardScalar help?

#### Training and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(calhouse['data'], calhouse['target'])
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

#### Linear Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
regr = linear_model.LinearRegression()
#train model
regr.fit(X_train, y_train)
#predicting with model
y_pred = regr.predict(X_test)
print("Coefficients: {}".format(regr.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred)))
print("Variance score: {}".format(r2_score(y_test, y_pred)))
```

#### Ridge

```python
from sklearn.linear_model import Ridge
clf = Ridge()
clf.fit(X_train, y_train)
y_pred_ridge = clf.predict(X_test)
print("Coefficients: {}".format(clf.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_ridge)))
print("Variance score: {}".format(r2_score(y_test, y_pred_ridge)))
```

#### Lasso

```python
from sklearn.linear_model import Lasso
las = Lasso()
las.fit(X_train, y_train)
y_pred_lasso = las.predict(X_test)
print("Coefficients: {}".format(las.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_lasso)))
print("Variance score: {}".format(r2_score(y_test, y_pred_lasso)))
```

#### ElasticNet

```python
from sklearn.linear_model import ElasticNet
EN = ElasticNet()
EN.fit(X_train, y_train)
y_pred_EN = EN.predict(X_test)
print("Coefficients: {}".format(EN.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_EN)))
print("Variance score: {}".format(r2_score(y_test, y_pred_EN)))
```

#### Scaling with StandardScalar

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(calhouse['data'], calhouse['target'])
scaledata = scaler.transform(calhouse['data'], calhouse['target'])
print("Scaler mean: {}".format(scaler.mean_))
X_trainscale, X_testscale, y_trainscale, y_testscale = train_test_split(scaledata, calhouse['target'])
print("\nLinear Regression")
regr.fit(X_trainscale, y_trainscale)
y_predscale = regr.predict(X_testscale)
print("Coefficients: {}".format(regr.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_testscale, y_predscale)))
print("Variance score: {}".format(r2_score(y_testscale, y_predscale)))
print("\nRidge")
clf.fit(X_trainscale, y_trainscale)
y_predscale_ridge = clf.predict(X_testscale)
print("Coefficients: {}".format(regr.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_testscale, y_predscale_ridge)))
print("Variance score: {}".format(r2_score(y_testscale, y_predscale_ridge)))
print("\nLasso")
las.fit(X_trainscale, y_trainscale)
y_predscale_lasso = las.predict(X_testscale)
print("Coefficients: {}".format(regr.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_testscale, y_predscale_lasso)))
print("Variance score: {}".format(r2_score(y_testscale, y_predscale_lasso)))
print("\nElasticNet")
EN.fit(X_trainscale, y_trainscale)
y_predscale_EN = EN.predict(X_testscale)
print("Coefficients: {}".format(regr.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_testscale, y_predscale_EN)))
print("Variance score: {}".format(r2_score(y_testscale, y_predscale_EN)))
```

Scaling the data with StandardScaler improved the results for Linear Regression and Ridge, but made the results for Lasso and ElasticNet much worse.

### d) Tune the parameters of the models using GridSearchCV.  Do the results improve?  Visualize the dependence of the validation score on the parameters for Ridge, Lasso, and ElasticNet.

##### Lasso

```python
from sklearn.model_selection import GridSearchCV
parameters = {'alpha':[0.0005,0.001,0.0015]}
CV = GridSearchCV(las, parameters, cv=5)
CV.fit(calhouse.data, calhouse.target)
print("Best parameter found: {}".format(CV.best_params_))
print("Means: {}".format(CV.cv_results_['mean_test_score']))
print("Standard Deviations: {}".format(CV.cv_results_['std_test_score']))
print("\nOriginal")
las = Lasso()
las.fit(X_train, y_train)
y_pred_lasso = las.predict(X_test)
print("Coefficients: {}".format(las.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_lasso)))
print("Variance score: {}".format(r2_score(y_test, y_pred_lasso)))
print("\nNew")
las2 = Lasso(alpha=0.001)
las2.fit(X_train, y_train)
y_pred_lasso2 = las2.predict(X_test)
print("Coefficients: {}".format(las2.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_lasso2)))
print("Variance score: {}".format(r2_score(y_test, y_pred_lasso2)))
```

##### Ridge

```python
from sklearn.model_selection import GridSearchCV
parameters = {'alpha':[82,81,80]}
CV = GridSearchCV(clf, parameters, cv=5)
CV.fit(calhouse.data, calhouse.target)
print("Best parameter found: {}".format(CV.best_params_))
print("Means: {}".format(CV.cv_results_['mean_test_score']))
print("Standard Deviations: {}".format(CV.cv_results_['std_test_score']))
print("\nOriginal")
clf = Ridge()
clf.fit(X_train, y_train)
y_pred_ridge = clf.predict(X_test)
print("Coefficients: {}".format(clf.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_ridge)))
print("Variance score: {}".format(r2_score(y_test, y_pred_ridge)))
print("\nNew")
clf2 = Ridge(alpha=81)
clf2.fit(X_train, y_train)
y_pred_ridge2 = clf2.predict(X_test)
print("Coefficients: {}".format(clf2.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_ridge2)))
print("Variance score: {}".format(r2_score(y_test, y_pred_ridge2)))
```

##### ElasticNet

```python
from sklearn.model_selection import GridSearchCV
parameters = {'alpha':[0.002,0.0019,0.0021]}
CV = GridSearchCV(EN, parameters, cv=5)
CV.fit(calhouse.data, calhouse.target)
print("Best parameter found: {}".format(CV.best_params_))
print("Means: {}".format(CV.cv_results_['mean_test_score']))
print("Standard Deviations: {}".format(CV.cv_results_['std_test_score']))
print("\nOriginal")
EN = ElasticNet()
EN.fit(X_train, y_train)
y_pred_EN = EN.predict(X_test)
print("Coefficients: {}".format(EN.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_EN)))
print("Variance score: {}".format(r2_score(y_test, y_pred_EN)))
print("\nNew")
EN2 = ElasticNet(alpha=0.0019)
EN2.fit(X_train, y_train)
y_pred_EN2 = EN2.predict(X_test)
print("Coefficients: {}".format(EN2.coef_))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_EN2)))
print("Variance score: {}".format(r2_score(y_test, y_pred_EN2)))
```

When tuning the parameters of ElasticNet, Ridge, and Lasso, Lasso was the program that improved the most. Lasso's variance score went from 0.3 to 0.6 and the mean squared value went from 0.95 to 0.55.  Ridge barely changed at all - the scores didn't improvve by more than 0.1.  ElasticNet improved somewhat.  The mean squared score went from 0.77 to 0.55 and the variance score went from 0.42 to 0.59.
The lower the mean squared score and the higher the variance score, the better the estimation fits the data.
